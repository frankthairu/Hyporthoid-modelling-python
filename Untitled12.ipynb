# Defining Data Question



## a) Specifying the Question

We are building  a model that determines whether or not the patient's symptoms indicate that the patient has hypothyroid

## b) Defining the Metric for success.

The model we will consider is the one that is able to effectively and optimally predict the target variable with 60% to 95%

## c) Understanding the context

The dataset that we will be using is  Nairobi hospital thyroid dataset with the following columns:

column description

status - tells us if the patient is negative or has hypothyroidism

age - how old is the patient

sex - if the patient is male or female

on_thyroxine - true /False

query_on_thyroxine -true / false

on_antithyroid_medication - if the patient is on medication(true / false)

thyroid_surgery - if the patient has had thyroid surgery(true / false)

query_hypothyroid-true / false

query_hyperthyroid-true / false

pregnant - if the patient is pregnant,true or false(true / false)

sick - if the patient is sick,(true / false)

tumor - if the patient has tumor(true / false)

lithium-(true / false)

goitre - if the patient has goitre(true / false)

TSH_measured - if a blood test has been done to measure the hormone(yes/no)

TSH - thyroid stimulating hormone , tells the thyroid to release thyroid hormone to the blood.

T3_measured - if T3 measurement has been done(yes/no)

T3 - triiodothyronine , this is a thyroid hormone produced in the thyroid glands

TT4_measured -if TT4 measurement has been done(yes/no)

TT4 - thyroxine , the principla hormone from the thyroid gland.

T4U_measured - if T4u measurement has been done(yes/no)

T4U - thyroid hormone produced in the thyroid glands(thyroxine)/levothyroxine

FTI_measured - if FTI has been measured(yes/no)

FTI - free thyroxine index, it is considered more indicator of thyroid, obtained by nultiplying Total T4 and T3 uptake.

TBG_measured - if TBG has been measured.(yes/no)

TBG - thyroid binding globulin, binds thyroid hormones in circulation

# Data cleaning

#### Changing columns to lowercase

```python
# Change all column names to lower case
df.columns=df.columns.str.lower()
df.columns
```

```python
# Checking if there are missing values 
df.isnull().sum()
```

At first there are no null values but there ar question marks(?) which we will remove below.

```python
# We check records weith (?) and replace them with null 

question_mark = ['age', 'sex', 'tsh', 't3', 'tt4', 't4u', 'fti', 'tbg']

for column in question_mark:
    df[column] = df[column].replace('?', np.NaN)

# Checking to see whether the null values are still present

print(df.isna().sum())



```

we have replaced the(?) marks with null values

#### Converting data types

```python
#coverting data from string to numeric
df[['age','tsh','t3','tt4','t4u','fti','tbg']] = df[['age','tsh','t3','tt4','t4u','fti','tbg']].apply(pd.to_numeric)
df.dtypes
```

```python
#We convert the object datatypes in our columns to category datatype
df[['status', 'sex', 'on_thyroxine', 'query_on_thyroxine','on_antithyroid_medication','thyroid_surgery','query_hypothyroid','query_hyperthyroid','pregnant','sick','tumor','lithium','goitre']] = df[['status', 'sex', 'on_thyroxine', 'query_on_thyroxine','on_antithyroid_medication','thyroid_surgery','query_hypothyroid','query_hyperthyroid','pregnant','sick','tumor','lithium','goitre']].astype('category')
df.dtypes
```

We convert the data types to appropriate datatypes in order to perform our EDA.

We also convert the datatypes in order for us to succesfully encode categorical columns.

#### Dealing with null values

```python
# We  fill the missing values in our columns 
# For our numerical columns we will fill the null values with the mean for each column

columns = ['age', 'tsh', 't3', 'tt4', 't4u', 'fti', 'tbg']

for column in columns:
    df[column] = df[column].fillna(df[column].mean())

# for the sex column, we will fill the null values with zero

df['sex'] = df['sex'].fillna(df['sex'].mode()[0])

# checking to see if the null are replaced
df.isna().sum()
```

We replace the null values in our numerical columns with mean. This method yields better results especially when our data is normally distributes as we will see below.

For the categorical column, we replece with mode since we can input the mean.

#### Dropping unnecessary columns

```python
#Cleaning our data
#dropping column that we dont need
df.drop(['tsh_measured', 't3_measured', 'tt4_measured', 't4u_measured', 'fti_measured', 'tbg_measured',], axis = 1, inplace = True)
df.head()
```

We drop these columns since they act as repetition to the columns next to them.

# Random Forest

```python
#We first change columns into categorical columns
df1[['status', 'sex', 'on_thyroxine', 'query_on_thyroxine','on_antithyroid_medication','thyroid_surgery','query_hypothyroid','query_hyperthyroid','pregnant','sick','tumor','lithium','goitre']] = df1[['status', 'sex', 'on_thyroxine', 'query_on_thyroxine','on_antithyroid_medication','thyroid_surgery','query_hypothyroid','query_hyperthyroid','pregnant','sick','tumor','lithium','goitre']].astype('category')
df1.dtypes
```

```python
# To create our model, we first split our data:

# Splitting the data into Features set and Labels

X= df1.drop('status', axis = 1)


# Getting dummies for the feature sets

X = pd.get_dummies(X)

# Encoding the target variable from strings to numericals
           
y = df1.status.cat.codes

# Splitting the data again into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
 
# Fitting the model, making predictions and evaluating the model

forest = RandomForestClassifier(n_estimators = 100, random_state=42, min_samples_split=10, max_depth=5)

forest = forest.fit(X_train, y_train)

y_pred = forest.predict(X_test)

accuracy_score(y_test, y_pred)
```

Our model is 97.8% which is noth that bad

```python
# Visualizing tree number 40
from IPython.display import Image
y_pred = forest.predict(X_test)
dot_data = StringIO()

# picking a specific tree from the forest; tree  number 40
tree = forest.estimators_[40]

export_graphviz(tree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = X.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Hypothyroid_Test.png')
Image(graph.create_png())
```
